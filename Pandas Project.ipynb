```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
gas = pd.read_csv('gas_prices.csv')
gas
```

```python
plt.figure(figsize=(15,5))
plt.title('Gas Prices per year in USD')

plt.xticks(gas.Year)

plt.ylabel('Price in USD')
plt.xlabel('Year')

plt.plot(gas['Year'], gas.USA, label = 'United States')
plt.plot(gas.Year, gas.Canada, label = 'Canada')
# Use [''] if two words - like South Korea
plt.plot(gas.Year, gas['South Korea'], label = 'South Korea')

plt.legend()
plt.show()
```

```python
#take every nth element of list - list[start:stop:step]

plt.figure(figsize=(8,5))
plt.title('Gas Prices per year in USD')

plt.xticks(gas.Year[::3])

plt.ylabel('Price in USD')
plt.xlabel('Year')

plt.plot(gas['Year'], gas.USA, 'b.-', label = 'United States')
plt.plot(gas.Year, gas.Canada, 'r.-', label = 'Canada')
# Use [''] if two words - like South Korea
plt.plot(gas.Year, gas['South Korea'], 'g.-', label = 'South Korea')

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title('Gas Prices per year in USD', fontdict = {'fontweight':'bold', 'fontsize':20})

for country in gas:
    if country != 'Year':
        plt.plot(gas.Year, gas[country], marker = '.', label = country)

plt.xticks(gas.Year[::3])
#plt.xticks(gas.Year[::3].tolist()+[2011])

plt.ylabel('Price in USD')
plt.xlabel('Year')
plt.legend()

plt.savefig('Gas prices per year of 10 Countries.png', dpi=300)
plt.show()
```
